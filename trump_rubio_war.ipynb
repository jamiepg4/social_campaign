```python
#open mongodb connection
from pymongo import MongoClient 
client = MongoClient()
```

```python
#import libraries
import datetime
import dateutil.relativedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ggplot import *
```

```python
#set options
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', -1)
```

```python
#get field names
client.scrape2.posts.find_one()
```

```python
#set start date
now = datetime.datetime.now()
start = now - dateutil.relativedelta.relativedelta(hours=72)
print start
```

```python
##pd.unique(df.candidate.ravel())
greenlist = ['rubio', 'trump']
```

```python
#query database
db = client.scrape2
input_data = db.posts
df = pd.DataFrame(list(input_data.find(
            {"$and": [{"created_at": {"$gte": start}}, {"candidate": {"$in": greenlist}}]},
            {"first_seen_at":0, "media": 0, "raw":0, "popularity_history":0} 
        )
                        ))
df.head(10)
```

## Process Data

```python
df.to_csv('2016_02_07_mostpopular_posts.csv', encoding='utf-8')
#pophist.to_csv('2016_02_07_phbycandidate.csv')
```

```python
#unnest nested fields 
#pull "popularity" field, reimport as dataframe
pop = pd.DataFrame(list(df.popularity))
pop.head(10)
```

```python
#cbind unnested "popularity" table with dataframe
flat_df = pd.concat([df, pop], axis=1)
flat_df.pop('popularity')
flat_df.head(1)
```

```python
#pull in labels from "Scraper Status" googlesheet 
scraper_status = pd.read_csv(
    'https://docs.google.com/spreadsheets/d/1tRlQhGQ9vY5N93n8WvN5fJ8t_SzH9xsxLquGiMfIALE/pub?gid=0&single=true&output=csv')
scraper_status = pd.DataFrame(scraper_status)
df_labels = scraper_status.loc[:,["Slug", "Name", "Party"]]
df_labels.head(3)
```

```python
#merge popularity df with labels 
pop_merged = pd.merge(left=flat_df, right=df_labels, how='left', left_on ='candidate', right_on ='Slug', indicator = True)
#merged_inner = pd.merge(left=surveySub,right=speciesSub, left_on='species_id', right_on='species_id')
pop_merged = pop_merged.drop_duplicates() #note dedupping is required because of a bug in Pandas. 
pop_merged.shape
#pop_merged.head(10)
```

```python
#convert categorical variables 
pop_merged['candidate'] = pop_merged['candidate'].astype('category')
pop_merged['Name'] = pop_merged['Name'].astype('category')
pop_merged['Slug'] = pop_merged['Slug'].astype('category')
pop_merged['Party'] = pop_merged['Party'].astype('category')
pop_merged['platform'] = pop_merged['platform'].astype('category')
pop_merged.info()
```

```python
by_date = pop_merged.set_index('created_at').sort_index(ascending = True)
by_date 
```
